```python
#ask number from the user . print all the numbers from 1 to that number
num = int(input("enter the number : "))

for i in range (1,num+1,1):
  print(i,end=" ")
```

```python
#ask start number and end number from the user and . print all the number from the start to end using while loop

start = int(input("Enter the number: "))
end = int(input("Enter the number : "))

for i in range (start,end+1,1):
  print(i,end=" ")
```

```python
# calculate the sum of all number from 1 to 10
sum=0
for i in range (1,11,1):
  sum=sum+i
print(sum)
```

```python
#claculate the product of all the numbers from 1 to 10

prod = 1
for i in range (1,11,1):
  prod=prod*i
print(prod)
```

```python
#calculate how many numbers are divisible by 7 from 1 to 100

count = 0
for i in range (1,101,1):
  if(i%7==0):
    count=count+1
print(count)
```
