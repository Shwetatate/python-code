```python
for i in range (1,11):
  if i==5:
    break
  print(i)
```

```python
for i in range (1,11):
  print(i)
  if (i==5):
    break
```

```python
for i in range(1,11):
  if i==5:
    continue
  print(i)
  print("Done")

print("program finish")
```

Nested loops

```python
for i in range (1,4):
   print(i)
   for j in range(10,14):
    print(j)
```

```python
"""
* * * * *
* * * * *
* * * * *
* * * * *
* * * * *
"""

for i in range (1,6):
  for j in range (1,6):
    print("*",end=" ")
  print()
```

```python
"""
12345
12345
12345
12345
12345
"""

for i in range (1,6):
  for j in range(1,6):
    print(j,end=" ")
  print()
```

```python
# ask N from the user . N means number of lines . then print the following pattern

N = int(input("Enter the number: "))

for i in range(1,N+1):
  for j in range(1,N+1):
    print(i,end=" ")

  print()
```
